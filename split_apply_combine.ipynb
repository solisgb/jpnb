```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
```

```python
gapminder = pd.read_csv(r'./gapminder/gapminder.csv')
```

```python
gapminder
```

```python
p = sns.FacetGrid(
    data=gapminder,
    hue='continent',
    col='continent',
    col_wrap=2,
    height=4,
    aspect=16/9
).map(plt.plot, 'year', 'lifeExp')
```

```python
models = pd.DataFrame({
    # this works because grouped dataframes in pandas are iterable
    # and because you can pretty much treat series objects like
    # they are lists
    'data': [data for _, data in gapminder.groupby('country')],
})
```
