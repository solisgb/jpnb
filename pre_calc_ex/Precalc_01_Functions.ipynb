```python
import numpy as np
import matplotlib.pyplot as plt
```

## Vertical and Horizontal Shifts

```python
f = lambda x: np.sqrt(x)
```

```python
x = np.linspace(0, 5, 25)

vs1 = 2.
vs2 = -vs1

fig, ax = plt.subplots(1,1)
ax.plot(x, f(x), label='f(x)')
ax.plot(x, f(x)+vs1, label=f'vs {vs1:0.1f}')
ax.plot(x, f(x)+vs2, label=f'vs {vs2:0.1f}')
ax.legend()
```

```python
f = lambda x: np.power(x, 2)
hs1 = 2.
hs2 = -hs1

start = -4; end = 4

fig, ax = plt.subplots(1,1)
x = np.linspace(start-hs1, end+hs1, 25)
ax.plot(x, f(x), label='f(x)')
x = np.linspace(start-hs1, end+hs1, 25)
ax.plot(x, f(x+hs1), label=f'hs {hs1:0.1f}')
ax.plot(x, f(x+hs2), label=f'hs {hs2:0.1f}')
ax.legend()
#plt.axis('square')
```

## Reflections
* Given a function f (x), a new function g(x) = −f (x) is a vertical reflection of the function f (x), sometimes called a reflection about (or over, or through) the x-axis.
* Given a function f (x), a new function g(x) = f (−x) is a horizontal reflection of the function f (x), sometimes called a reflection about the y-axis.

```python
# reflecting

f = lambda x: np.sqrt(x)

x = np.arange(0, 5, 0.1)
y = f(x)
x2 = np.arange(0, -5, -0.1)
y2 = f(-1*x2)

fig, ax = plt.subplots(1,1)
ax.plot(x, y)
ax.plot(x2, y2,label=f'y2')
ax.legend()
```

```python
# combination of various transformations

f = lambda x: np.exp2(x)

x = np.linspace(-4.5, 4.5, 20)
y = f(x)
# horizontal reflection
y2 = f(-x)
# vertical reflection
y3 = -y2
# vertical shift
y4 = y3 + 4
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(x, y, label='2^x')
ax.plot(x, y2, label='2^-x')
ax.plot(x, y3, label='-2^-x')
ax.plot(x, y4, label='-2^-x + 1', color='blue')
ax.legend()
```

```python
f = lambda x: np.power(x, 2)

x = np.linspace(-4.5, 4.5, 20)

fig, ax = plt.subplots(1,1)
ax.plot(x, f(x), label='x^2')
ax.plot(x, -f(x), label='-(x^2)')
ax.plot(x, f(-x), label='-x^2', linestyle=' ',marker='.', color='red')
ax.legend()
```

## Determining Even and Odd Functions

```python
print('Even function: If for every input x: f(x) = f(−x): its graph is symmetric about the y-axis')
print('Odd function: If for every input x: f(x) = −f(−x): its graph is symmetric about the origin')
print('f=(x^3) is an odd function')
x = np.linspace(-1.5, 1.5, 50)
fig, ax = plt.subplots(1,3, sharex=True)
ax[0].plot(x, np.power(x, 3), label='x^3')
ax[1].plot(x, np.power(-x, 3), label='-x^3')
ax[2].plot(x, -1 * np.power(-x, 3), label='-(-x^3)')
for i in range(len(ax)):
    ax[i].legend()
```

## Graphing Functions Using Stretches and Compressions

Given a function f (x), a new function g(x) = af(x), where a is a constant, is a vertical stretch or vertical compression of the function f (x)
* If a > 1, then the graph will be stretched.
* If 0 < a < 1, then the graph will be compressed.
* If a < 0, then there will be combination of a vertical stretch or compression with a vertical reflection

```python
f = lambda x: np.power(x, 2)

x = np.linspace(-5, 5, 50)
fig, ax = plt.subplots(figsize=(4,5))
ax.plot(x, np.power(x, 2), label='x^2')
ax.plot(x, np.power(2*x, 2), label='2x^2')
ax.plot(x, np.power(0.5*x, 2), label='0.5*x^2')
ax.legend()
```

```python
f = lambda x: np.sin(x)

x = np.linspace(-10, 10, 50)

fig, ax = plt.subplots()
ax.plot(x, f(x), label='sin(x)')
ax.plot(x, f(2*x), label='2f(x)')
ax.plot(x, f(0.5*x), label='0.5f(x)')
ax.legend()
```

## Combining transformations

1. When combining vertical transformations written in the form af(x) + k, first vertically stretch by a and then vertically shift by k.
1. When combining horizontal transformations written in the form f (bx − h), first horizontally shift by h and then horizontally stretch by 1/b.
1. When combining horizontal transformations written in the form f (b(x − h)), first horizontally stretch by 1/b and then horizontally shift by h.
1. Horizontal and vertical transformations are independent. It does not matter whether horizontal or vertical transformations are performed first.

$ y = \sqrt{r^2-x^2} $

$ f(\frac{1}{2}x+1)-3 ) = f(\frac{1}{2}(x+2))-3$

```python
f = lambda x, r: np.sqrt(r**2 - x**2)

r = 2
x = np.linspace(-r, r, 50)

fig, ax = plt.subplots()
ax.plot(x, f(x, r), label='f(x)')

x = np.linspace(-4, 4, 50)
y = f(0.5*x, r)
ax.plot(x, y, label='f(0.5*x, r)')

x = np.linspace(-6, 2, 49)
y = f(0.5*(x+2), r)
ax.plot(x, y, label='f(0.5*(x+2), r)')
ax.plot(x, y-3, label='f(0.5*(x+2)-3, r)')
ax.legend()
ax.grid(True)
ax.axis('equal')
```
